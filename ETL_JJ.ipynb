# Transform

```python
# Rename column headers for scc
state_commute_clean = scc_df.rename(columns={"state_name": "State",
                                   "commute_cost": "Average Annual Commute Cost"})
state_commute_clean.head()
```

```python
# Set Indices
state_commute_clean.set_index("State", inplace=True)
state_commute_clean.head()
```

```python
mc_df.set_index("State", inplace=True)
mc_df.head()
```

```python
tc_df.set_index("State", inplace=True)
tc_df.head()
```

```python
lc_df.index = [x for x in range(1, len(lc_df.values)+1)]
lc_df.head()
```

```python
mc2_df.index = [x for x in range(1, len(mc2_df.values)+1)]
mc2_df.head()
```

```python
sc_df.index = [x for x in range(1, len(sc_df.values)+1)]
sc_df.head()
```

```python
# Merge DataFrames
merged_data = pd.merge(pd.merge(state_commute_clean,mc_df,on='State'),tc_df,on='State')
merged_data.head()
```

```python
# Clean city bike data
```

```python
# Large City DF
lc_drop = lc_df.drop(lc_df.index[1])
lc_drop2 = lc_drop.drop(lc_drop.index[1])
lc_drop3 = lc_drop2.drop(lc_drop2.index[0])
lc_drop4 = lc_drop3.drop(lc_drop3.index[0])
lc_drop5 = lc_drop4.drop(lc_drop4.index[0])
lc_drop5.head()
```

```python
# Medium City DF
mc2_drop = mc2_df.drop(mc2_df.index[1])
mc2_drop2 = mc2_drop.drop(mc2_drop.index[1])
mc2_drop3 = mc2_drop2.drop(mc2_drop2.index[0])
mc2_drop4 = mc2_drop3.drop(mc2_drop3.index[0])
mc2_drop5 = mc2_drop4.drop(mc2_drop4.index[0])
mc2_drop5.head()
```

```python
# Small City DF
sc_drop = sc_df.drop(sc_df.index[1])
sc_drop2 = sc_drop.drop(sc_drop.index[1])
sc_drop3 = sc_drop2.drop(sc_drop2.index[0])
sc_drop4 = sc_drop3.drop(sc_drop3.index[0])
sc_drop5 = sc_drop4.drop(sc_drop4.index[0])
sc_drop5.head()
```

```python
# Rename Column Headers and reset index
```

```python
# Large Cities
lc_clean = lc_drop5.rename(columns={"Table with row headers in Column A, and column headers in row 6": "City",
                                   "Unnamed: 1": "Number of Workers who commuted by bicycle",
                                   "Unnamed: 2": "Percentage of workers",
                                   "Unnamed: 3": "Margin of \nerror (Â±)2"})
lc_clean.head()
```

```python
# Medium Cities
mc2_clean = mc2_drop5.rename(columns={"Table with row headers in column A, and Column headers in row 6": "City",
                                   "Unnamed: 1": "Number of Workers who commuted by bicycle",
                                   "Unnamed: 2": "Percentage of workers",
                                   "Unnamed: 3": "Margin of \nerror (Â±)2"})
mc2_clean.head()
```

```python
# Small Cities
sc_clean = sc_drop5.rename(columns={"Table with row headers in Column A, and column headers in row 3": "City",
                                   "Unnamed: 1": "Number of Workers who commuted by bicycle",
                                   "Unnamed: 2": "Percentage of workers",
                                   "Unnamed: 3": "Margin of \nerror (Â±)2"})
sc_clean.head()
```

```python
# Join city DFs
city_join = pd.concat([lc_clean,mc2_clean,sc_clean])
city_join
```

```python
city_clean = city_join.dropna()
city_clean
```

```python
city_clean.columns
```

```python
city_clean.set_index('City', inplace=True)
city_clean.head()
```

```python
city_clean.sort_values(by='City', ascending=True)
```

```python
# Rename State Data column names to match PostreSQL Table
state_data_final = merged_data.rename(columns={"State": "State",
                                   "Average Annual Commute Cost": "avg_annual_commute_cost",
                                   "Avg. Commute (mi)": "avg_commute_distance",
                                   "Avg. Commute Time (min)": "avg_commute_time"})
state_data_final.head()
```

```python
state_data_final2 = state_data_final.reset_index()
state_data_final2.head()
```

```python
# Rename City Data column names to match PostreSQL Table
city_data_final = city_clean.rename(columns={"City": "city",
                                   "Number of Workers who commuted by bicycle": "bike_commuters",
                                   "Percentage of workers": "percent_total_workforce",
                                   "Margin of \nerror (Â±)2": "error_margin"})
city_data_final.head()
```

```python
city_data_final2 = city_data_final.reset_index()
city_data_final2.head()
```
